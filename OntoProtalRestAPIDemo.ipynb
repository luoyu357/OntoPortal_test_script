````python
import urllib.request, urllib.error, urllib.parse
import json
import os
from pprint import pprint
from urllib.parse import quote
````

````python
REST_URL = "http://3.95.29.226:8080"
````

````python
API_KEY = "9b01b1cd-696d-4ef5-ac36-0b212f44599a"
````

````python
def get_json(url):
    opener = urllib.request.build_opener()
    opener.addheaders = [('Authorization', 'apikey token=' + API_KEY)]
    return json.loads(opener.open(url).read())
````

## List all ontologies

````python
# Get the available resources
resources = get_json(REST_URL + "/")

# Get the ontologies from the `ontologies` link
ontologies = get_json(resources["links"]["ontologies"])

# Print the first ontology in the list
pprint(ontologies[0])

# Get the name and ontology id from the returned list
ontology_output = []
for ontology in ontologies:
    ontology_output.append(f"{ontology['name']}\n{ontology['@id']}\n")


# Print the names and ids
print("\n\n")
for ont in ontology_output:
    print(ont)
````

## Get Terms

````python
ontologies = quote('IOT-STREAM')
result = get_json(REST_URL+"/ontologies/"+ontologies)
````

````python
terms = []
page = get_json(result["links"]["classes"])

# Iterate over the available pages adding labels from all classes
# When we hit the last page, the while loop will exit
next_page = page
while next_page:
    next_page = page["links"]["nextPage"]
    for bro_class in page["collection"]:
        terms.append(bro_class["prefLabel"])
    if next_page:
        page = get_json(next_page)

# Output the labels
for term in terms:
    print(term)
````

## Get Properties

````python
properties = []
page = get_json(result["links"]["properties"])
# Iterate over the available pages adding labels from all classes
# When we hit the last page, the while loop will exit

for bro_class in page:
    if bro_class["label"]:
        properties.append(bro_class["label"][0])


# Output the labels
for pro in properties:
    print(pro)
````
